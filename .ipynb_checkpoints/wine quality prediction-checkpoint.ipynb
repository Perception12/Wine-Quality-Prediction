# Wine Quality Prediction

## Importing the required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## Importing the datasets

```python
df1 = pd.read_csv('dataset/winequality-red.csv', delimiter=';')
df2 = pd.read_csv('dataset/winequality-white.csv', delimiter=';')

data = pd.concat([df1, df2])
```

## Exploratory Data Analysis

```python
grids = sns.FacetGrid(data, col="quality")
grids.map(sns.scatterplot, "total sulfur dioxide", "density", alpha=.8)
```

```python
grids = sns.FacetGrid(data, col="quality")
grids.map(sns.scatterplot, "fixed acidity", "pH", alpha=.8)
```

```python
fig, ax = plt.subplots(figsize=(10,10))  
sns.heatmap(data.corr(), annot=True)
```

## Getting dependent and independent variables

### For full features

```python
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

## Splitting data into train and test sets

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## Training our models to find the best one

### Linear Regression

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression() 
lin_reg.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
def print_metrics(pred):
    print("Full Data Metric\n")
    print("Mean absolute error: ", mean_absolute_error(y_test, pred))
    print("Mean squared error: ", mean_squared_error(y_test, pred))
    print("r2 score: ", r2_score(y_test, pred))
```

```python
print_metrics(lin_reg.predict(X_test))
```

### Support Vector Regression

```python
from sklearn.svm import SVR 

svr_reg = SVR(kernel='rbf')

svr_reg.fit(X_train, y_train)
```

```python
print_metrics(svr_reg.predict(X_test))
```

### Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor

dtr_reg = DecisionTreeRegressor(max_depth=6)
dtr_reg.fit(X_train, y_train)
```

```python
print_metrics(dtr_reg.predict(X_test))
```

### Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor

rf_reg = RandomForestRegressor(n_estimators=20)
rf_reg.fit(X_train, y_train)
```

```python
print_metrics(rf_reg.predict(X_test))
```
